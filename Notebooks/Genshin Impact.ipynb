# Gráficos

# Número de Personagens por Elemento

```python
element = genshin_df['Element'].value_counts()
total = sum(element)

element.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * total / 100))
plt.ylabel("Count", size=0)
plt.title("Número de Personagens por Elemento", size=18)
plt.show()
```

# Número de Personagens por Arma

```python
weapon = genshin_df['Weapon'].value_counts()
total = sum(weapon)

weapon.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * total / 100))
plt.ylabel("Count", size=0)
plt.title("Número de Personagens por Arma", size=18)
plt.show()
```

# Número de Personagens por Main Role

```python
mainrole = genshin_df['MainRole'].value_counts()
total = sum(mainrole)

mainrole.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * total / 100))
plt.ylabel("Count", size=0)
plt.title("Número de Personagens por Main Role", size=18)
plt.show()
```

# Tipo de Main Role por Elemento

```python
me = genshin_df.groupby(['Element','MainRole']).size().unstack()

me.plot(kind = 'barh', stacked = True, figsize = [12,8])

plt.title("Tipo de Main Role por Elemento", size=18)
plt.xlabel("Valores", size=12)
plt.ylabel("Elementos", size=12)
plt.show()
```

# Comparação entre Personagens raridade 4 e 5

```python
genshin4_df=genshin_df.loc[(genshin_df.Rarity==4), :]
genshin4_df
```

```python
genshin5_df=genshin_df.loc[(genshin_df.Rarity==5), :]
genshin5_df
```

```python
Characters = genshin4_df['Character']
BHP = genshin4_df['BHP']
BATK = genshin4_df['BATK']
BDEF = genshin4_df['BDEF']

fig, ax = plt.subplots(figsize = [12,8])

ax.bar(Characters, BHP)
ax.bar(Characters, BATK, bottom=BHP)
ax.bar(Characters, BDEF, bottom=BHP + BATK)

plt.title("Personagens de Raridade 4", size=18)
plt.xlabel("Personagens", size=12)
plt.ylabel("Valores", size=12)
plt.xticks(rotation=80)
plt.show()
```
