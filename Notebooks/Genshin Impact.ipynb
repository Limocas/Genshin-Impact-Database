```python
import matplotlib.pyplot as plt
import pandas as pd

genshin = pd.read_csv('Genshin_Impact_All_Character_Stat.csv')
```

# Gestão dos dados

```python
genshin.head()
```

```python
genshin.tail()
```

```python
# Mudar os nomes de certas colunas para não haver problemas com espaços
genshin = genshin.rename(columns = {'Main role': 'MainRole','Base HP': 'BHP', 'Base ATK': 'BATK', 'Base DEF': 'BDEF'})
```

```python
# Filtrar o dataframe de forma a ser só personagens de nível 90, pois não é necessário comparar todos os níveis
genshin_df = genshin.loc[(genshin.Lv==90)]
genshin_df
```

```python
# Verificar que tipo de dados é cada coluna
genshin_df.dtypes
```

# Questões

__Quantas Personagens tem o jogo?__

```python
len(genshin_df['Character'])
```

__Quantas personagens são de raridade 4 e quantas de raridade 5?__

```python
genshin_df['Rarity'].value_counts()
```

__Quantas personagens há por elemento?__

```python
# Guardamos este valor numa variavél para podermos também fazer um gráfico sobre isto
element = genshin_df['Element'].value_counts()
element
```

```python
# Esta variável "totale" serve para podermos obter os valores certos para podermos observá-los no gráfico
totale = sum(element)

element.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * totale / 100))
plt.ylabel("", size=0)
plt.title("Número de Personagens por Elemento", size=18)
plt.show()
```

__Quantas personagens há por arma?__

```python
# Guardamos este valor numa variavél para podermos também fazer um gráfico sobre isto
weapon = genshin_df['Weapon'].value_counts()
weapon
```

```python
# Esta variável "totalw" serve para podermos obter os valores certos para podermos observá-los no gráfico
totalw = sum(weapon)

weapon.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * totalw / 100))
plt.ylabel("", size=0)
plt.title("Número de Personagens por Arma", size=18)
plt.show()
```

__Quantas personagens há por Main Role?__

```python
# Guardamos este valor numa variavél para podermos também fazer um gráfico sobre isto
mainrole = genshin_df['MainRole'].value_counts()
mainrole
```

```python
# Esta variável "totalm" serve para podermos obter os valores certos para podermos observá-los no gráfico
totalm = sum(mainrole)

mainrole.plot(kind='pie', figsize=(12, 8), autopct=lambda p: '{:.0f}'.format(p * totalm / 100))
plt.ylabel("", size=0)
plt.title("Número de Personagens por Main Role", size=18)
plt.show()
```

__Personagem com o maior Base DEF__

```python
maiorBDEF=genshin_df['BDEF'].max()
genshin_df[['Character','BDEF']][genshin_df.BDEF==maiorBDEF]
```

__Personagem com os Stats no total mais elevados__

```python
# Para esta questão, primeiro temos de ver quais são os stats no total mais elevados
BHP = genshin_df['BHP']
BATK = genshin_df['BATK']
BDEF = genshin_df['BDEF']

BALL = (BHP + BATK + BDEF)
# Aqui podemos então pedir ao programa o "BALL" mais elevado
maiorBALL=BALL.max()
print(maiorBALL)

# Assim podemos comparar todos os "BALL" com o maior que obtivemos em cima para obtermos a personagem que tem esse valor
genshin_df[['Character']][BALL==maiorBALL]
```

__Tipo de Main Role por Elemento__

```python
# Começamos por fazer groupby nos dois elementos em questão para os podermos usar num gráfico
me = genshin_df.groupby(['Element','MainRole']).size().unstack()

# De seguida, escolhemos o tipo de gráfico e "stacked = True" para os gráficos ficaram todos juntos
me.plot(kind = 'barh', stacked = True, figsize = [12,8])

plt.title("Tipo de Main Role por Elemento", size=18)
plt.xlabel("Valores", size=12)
plt.ylabel("Elementos", size=12)
plt.show()
```

__Comparação entre Personagens raridade 4 e 5__

```python
#Para este tipo de questão começamos por separar o dataframe em 2 dataframes para cada tipo de raridade
genshin4_df = genshin_df.loc[(genshin_df.Rarity==4)]
genshin4_df
```

```python
genshin5_df = genshin_df.loc[(genshin_df.Rarity==5)]
genshin5_df
```

__Gráfico com valores de ataque, vida e defesa por personagem de raridade 4__

```python
# Começamos por criar variáveis para cada parte do gráfico
Characters4 = genshin4_df['Character']
BHP4 = genshin4_df['BHP']
BATK4 = genshin4_df['BATK']
BDEF4 = genshin4_df['BDEF']

# Criar os gráficos
fig, ax = plt.subplots(figsize = [12,8])

# Aqui colocamos cada x e cada y e temos o "bottom" para indicar que ficará acima de um outro valor
ax.bar(Characters4, BHP4, label='Base HP')
ax.bar(Characters4, BATK4, bottom=BHP4, label='Base ATK')
ax.bar(Characters4, BDEF4, bottom=BHP4 + BATK4, label='Base DEF')

plt.title("Personagens de Raridade 4", size=18)
plt.xlabel("Personagens", size=12)
plt.ylabel("Valores", size=12)
plt.legend()
plt.xticks(rotation=80)
plt.show()
```

```python
maiorBHP4 = genshin4_df['BHP'].max()
genshin4_df[['Character','BHP']][genshin4_df.BHP==maiorBHP4]
```
